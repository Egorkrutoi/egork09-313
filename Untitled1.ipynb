```python
class Exhibit:
    def __init__(self, name):
        self.__name = name

    def get_name(self):
        return self.__name


class Museum:
    def __init__(self):
        self.__exhibits = []

    def add_exhibit(self, exhibit):
        if isinstance(exhibit, Exhibit):
            self.__exhibits.append(exhibit)
            print(f" '{exhibit.get_name()}'   ")
        else:
            print(":      Exhibit")

    def show_exhibits(self):
        if not self.__exhibits:
            print("    ")
            return

        print("  :")
        for i, exhibit in enumerate(self.__exhibits, 1):
            print(f"{i}. {exhibit.get_name()}")



print(" ")
louvre = Museum()

print("\n :")
louvre.add_exhibit(Exhibit(" "))
louvre.add_exhibit(Exhibit(" "))
louvre.add_exhibit("")

print("\n :")
louvre.show_exhibits()
```

```python
class Card:
    def __init__(self, number, limit):
        self.number = number
        self.limit = limit

    def __str__(self):
        return f" {self.number} (: {self.limit})"


class Client:
    def __init__(self, name):
        self.name = name
        self.cards = []

    def add_card(self, card):
        """   """
        self.cards.append(card)
        print(f" {self.name}  {card}")

    def show_cards(self):
        """    """
        print(f"\n  {self.name}:")
        if not self.cards:
            print(" ")
        for i, card in enumerate(self.cards, 1):
            print(f"{i}. {card}")


#  
print(" ")
client = Client(" ")

print("\n ")
card1 = Card("2282312313123", 50000)
card2 = Card("555116", 1000000)

print("\n  ")
client.add_card(card1)
client.add_card(card2)

print("\n  ")
client.show_cards()

print("\n     ")
print(f" : {card1}")
```

```python
from abc import ABC, abstractmethod

class CarPart(ABC):
    @abstractmethod
    def replace(self):
        pass


class Engine(CarPart):
    def replace(self):
        return " "


class Wheel(CarPart):
    def replace(self):
        return " "

print("  ")

engine = Engine()
print("\n  :")
print(engine.replace())

wheel = Wheel()
print("\n  :")
print(wheel.replace())
```

```python
from abc import ABC, abstractmethod
class IBook(ABC):
    @abstractmethod
    def update_author(self, author):
        pass

    @abstractmethod
    def get_book_info(self):
        pass


class Book(IBook):
    def __init__(self, title, author, isbn):
        self.title = title
        self.author = author
        self.isbn = isbn

    def update_author(self, new_author):
        if not isinstance(new_author, str) or not new_author.strip():
            raise ValueError("     ")
        self.author = new_author.strip()

    def get_book_info(self):
        return f": {self.title}\n: {self.author}\nISBN: {self.isbn}"


if __name__ == "__main__":
    print(" ")
    book = Book("", "", "234567")

    print("\n  :")
    print(book.get_book_info())

    print("\n ")
    book.update_author("")

    print("\n :")
    print(book.get_book_info())

    try:
        print("\n  ")
        book.update_author("")
    except ValueError as e:
        print(f": {e}")
```

```python
class Book:
    def getGenre(self):
        return "Unknown genre"

class FictionBook(Book):
    def getGenre(self):
        return "Fiction"

class NonFictionBook(Book):
    def getGenre(self):
        return "Non-Fiction"

book1 = FictionBook()
print(book1.getGenre())

book2 = NonFictionBook()
print(book2.getGenre())

book3 = Book()
print(book3.getGenre())
```
