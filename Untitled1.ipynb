```python
class Exhibit:
    def __init__(self, name):
        self.__name = name

    def get_name(self):
        return self.__name


class Museum:
    def __init__(self):
        self.__exhibits = []

    def add_exhibit(self, exhibit):
        if isinstance(exhibit, Exhibit):
            self.__exhibits.append(exhibit)
            print(f"Экспонат '{exhibit.get_name()}' добавлен в музей")
        else:
            print("Ошибка: можно добавлять только объекты класса Exhibit")

    def show_exhibits(self):
        if not self.__exhibits:
            print("В музее пока нет экспонатов")
            return

        print("Экспонаты в музее:")
        for i, exhibit in enumerate(self.__exhibits, 1):
            print(f"{i}. {exhibit.get_name()}")



print("Создаем музей")
louvre = Museum()

print("\nДобавляем экспонаты:")
louvre.add_exhibit(Exhibit("Мона Лиза"))
louvre.add_exhibit(Exhibit("Венера Милосская"))
louvre.add_exhibit("Картина")

print("\nПросмотр экспонатов:")
louvre.show_exhibits()
```

```python
class Card:
    def __init__(self, number, limit):
        self.number = number
        self.limit = limit

    def __str__(self):
        return f"Карта {self.number} (лимит: {self.limit})"


class Client:
    def __init__(self, name):
        self.name = name
        self.cards = []

    def add_card(self, card):
        """Добавляет новую карту клиенту"""
        self.cards.append(card)
        print(f"Клиенту {self.name} добавлена {card}")

    def show_cards(self):
        """Выводит список всех карт клиента"""
        print(f"\nКарты клиента {self.name}:")
        if not self.cards:
            print("Нет карт")
        for i, card in enumerate(self.cards, 1):
            print(f"{i}. {card}")


# Пример использования
print("Создаем клиента")
client = Client("Иван Петров")

print("\nСоздаем карты")
card1 = Card("2282312313123", 50000)
card2 = Card("555вор116", 1000000)

print("\nДобавляем карты клиенту")
client.add_card(card1)
client.add_card(card2)

print("\nПоказываем карты клиента")
client.show_cards()

print("\nКарта может существовать отдельно от клиента")
print(f"Отдельная карта: {card1}")
```

```python
from abc import ABC, abstractmethod

class CarPart(ABC):
    @abstractmethod
    def replace(self):
        pass


class Engine(CarPart):
    def replace(self):
        return "Замена двигателя"


class Wheel(CarPart):
    def replace(self):
        return "Замена колеса"

print("Демонстрация работы ")

engine = Engine()
print("\nПроцесс замены двигателя:")
print(engine.replace())

wheel = Wheel()
print("\nПроцесс замены колеса:")
print(wheel.replace())
```

```python
from abc import ABC, abstractmethod
class IBook(ABC):
    @abstractmethod
    def update_author(self, author):
        pass

    @abstractmethod
    def get_book_info(self):
        pass


class Book(IBook):
    def __init__(self, title, author, isbn):
        self.title = title
        self.author = author
        self.isbn = isbn

    def update_author(self, new_author):
        if not isinstance(new_author, str) or not new_author.strip():
            raise ValueError("Имя автора должно быть непустой строкой")
        self.author = new_author.strip()

    def get_book_info(self):
        return f"Название: {self.title}\nАвтор: {self.author}\nISBN: {self.isbn}"


if __name__ == "__main__":
    print("Создаем книгу")
    book = Book("Лучший", "Егор", "234567")

    print("\nИнформация о книге:")
    print(book.get_book_info())

    print("\nОбновляем автора")
    book.update_author("Егор")

    print("\nОбновленная информация:")
    print(book.get_book_info())

    try:
        print("\nПытаемся установить ")
        book.update_author("")
    except ValueError as e:
        print(f"Ошибка: {e}")
```
