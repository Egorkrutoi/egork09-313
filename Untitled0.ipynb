```python
def are_triangles_similar(a, b, c, d):
    ratio1 = max(a, b) / min(a, b)
    ratio2 = max(c, d) / min(c, d)
    return abs(ratio1 - ratio2) < 1e-9
```

```python
def find_unique_number(a1, a2, a3, a4):
    if a1 == a2 == a3:
        return 4
    elif a1 == a2 == a4:
        return 3
    elif a1 == a3 == a4:
        return 2
    else:
        return 1
```

```python
def analyze_number(number):
    is_even = number % 2 == 0
    ends_with_7 = number % 10 == 7
    return is_even, ends_with_7
```

```python
def can_white_move_safely(a, b, c, d, e, f, white_piece, black_piece):
    if not all(1 <= coord <= 8 for coord in [a, b, c, d, e, f]):
        return False

    def can_white_reach():
        if white_piece == 'ладья':
            return e == a or f == b
        elif white_piece == 'ферзь':
            return e == a or f == b or abs(e - a) == abs(f - b)
        else:
            return False
    def is_under_attack():
        if black_piece == 'ладья':
            return e == c or f == d
        elif black_piece == 'слон':
            return abs(e - c) == abs(f - d)
        elif black_piece == 'ферзь':
            return e == c or f == d or abs(e - c) == abs(f - d)
        else:
            return False

    return can_white_reach() and not is_under_attack()
```

```python
import math

def calculate_sum(N):
    total = 0.0
    for i in range(1, N + 1):
        denominator = sum(math.sin(j) for j in range(1, i + 1))
        total += 1 / denominator
    return total

print(calculate_sum(5))  # Пример для N = 5
```

```python
def sum_with_precision():
    epsilon = 0.001
    total = 0.0
    n = 1
    while True:
        term = (1 / (2 ** n)) + (1 / (3 ** n))
        total += term
        if term < epsilon:
            break
        n += 1
    return total

print(sum_with_precision())
```

```python
def last_number_greater_than_K(K, numbers):
    last_index = 0
    for i, num in enumerate(numbers, start=1):
        if num > K:
            last_index = i
    return last_index

# Пример использования:
numbers = [5, -3, 8, 12, 0]  # 0 — признак завершения
K = 6
print(last_number_greater_than_K(K, numbers))
```

```python
from fractions import Fraction

def sequence_k(k):
    a = Fraction(1, 1)
    for _ in range(1, k):
        a += Fraction(1, 1 + a.numerator // a.denominator)
    return a

k = 3
print(f"a_{k} = {sequence_k(k)}")
```

```python
import math

def distance_matrix(points):
    n = len(points)
    matrix = [[0.0] * (n - 1) for _ in range(n)]
    for i in range(n):
        others = [j for j in range(n) if j != i]
        for idx, j in enumerate(others):
            dx = points[i][0] - points[j][0]
            dy = points[i][1] - points[j][1]
            matrix[i][idx] = math.sqrt(dx**2 + dy**2)
    return matrix

points = [(0, 0), (1, 0), (0, 1)]
print(distance_matrix(points))
```

```python
def days_to_exceed():
    eaten = 12
    days = 1
    while eaten <= 100:
        days += 1
        eaten = eaten * 1.2 + 2
    return days

print(days_to_exceed())
```
