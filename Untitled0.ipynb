```python

def vch(a,b,c,d):
  return  ((a/c)*(b/d)-(a*b-c/c*d))
a = int(input())
b = int(input())
c = int(input())
d = int(input())
print(vch(a,b,c,d))
```

```python
from math import log
def asd(x):
  return ((log(x)**2/x)+log(x**2)/(1+log(x)**2))
x = int(input())

print(asd(x))
```

```python
from math import pi
def ring(r,R):
  return (pi*(R*R-r*2))
R = int(input())
r = int(input())

print(ring(R,r))
```

```python
def sut(N):
  return (N//60)
N = int(input())

print(sut(N))
```

```python
def ch(A,B,C):
  return (B>A and B<C)
A = int(input())
B = int(input())
C = int(input())

print(ch(A,B,C))
```
